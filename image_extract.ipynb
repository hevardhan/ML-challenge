```python
import os
import pandas as pd
import requests
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# Initialize Selenium WebDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Load your DataFrame (assuming CSV)
df = pd.read_csv('image_links.csv')  # Adjust this path as necessary

# Create a folder to save images
save_folder = "downloaded_images"
os.makedirs(save_folder, exist_ok=True)

# Iterate over the DataFrame and download images
for idx, row in df.iterrows():
    img_url = row['image_link']  # Column name should match your DataFrame
    driver.get(img_url)  # Open image link using Selenium to ensure accessibility
    
    # Request and save the image
    response = requests.get(img_url)
    if response.status_code == 200:
        file_name = os.path.join(save_folder, f'image_{idx}.jpg')
        with open(file_name, 'wb') as file:
            file.write(response.content)
        print(f"Image {idx} downloaded: {file_name}")
    else:
        print(f"Failed to download image {idx}")

# Quit the Selenium driver
driver.quit()
```

```python
import pandas as pd
import requests
from PIL import Image
from PIL import PngImagePlugin
import os
from io import BytesIO


# Create a directory to store images if it doesn't exist
if not os.path.exists('images_with_metadata'):
    os.makedirs('images_with_metadata')

# Function to download and save the image with custom metadata
def download_and_save_image_with_metadata(url, metadata, filename):
    try:
        # Download the image
        response = requests.get(url)
        if response.status_code == 200:
            # Open the image
            image = Image.open(BytesIO(response.content))

            # Prepare metadata (only works for PNG; EXIF data for JPEG is more restricted)
            meta = PngImagePlugin.PngInfo()
            for key, value in metadata.items():
                meta.add_text(key, str(value))

            # Save image with metadata
            image.save(filename, pnginfo=meta)
            print(f'Successfully downloaded and saved {filename} with metadata')

        else:
            print(f'Failed to download {url}')
    except Exception as e:
        print(f'Error downloading {url}: {e}')

# Iterate over the dataset and download images with metadata
for index, row in df.iterrows():
    image_url = row['image_link']  # URL of the image
    metadata = {
        "group_id": row['group_id'],  # Custom metadata
        "entity_name": row['entity_name'],
        "entity_value": row['entity_value']
    }
    filename = f"images_with_metadata/image_{index}.png"  # Save as PNG for metadata support
    download_and_save_image_with_metadata(image_url, metadata, filename)
```
